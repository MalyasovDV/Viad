```python
import numpy as np # linear algebra
import pandas as pd # data processing
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
n_cpu = os.cpu_count()
n_thread = n_cpu*2
```

```python
df=pd.read_csv("vehicles.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df= df.drop(columns=['id','url', 'region_url', 'VIN', 'image_url', 'description', 'lat', 'long','county','region','posting_date','state','size'], axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
for i in df:
    print(df[i].unique())
```

```python
#df.dropna(axis=0, how="any", inplace=True)
df = df.dropna(subset=['year', 'odometer'])
df.fillna('n/a', inplace=True)
```

```python
df.shape
```

```python
p1 = df['price'].quantile(0.99)
p2 = df['price'].quantile(0.1)
df = df[(df['price']<p1) & (df['price']>p2)]
o1 = df['odometer'].quantile(0.99)
o2 = df['odometer'].quantile(0.05)
df = df[(df['odometer']<o1) & (df['odometer']>o2)]
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df.columns
```

```python
X_num = df[['odometer', 'year']]
X_cat = df.drop(['odometer', 'year', 'price'], axis=1)
```

```python
for i in X_cat.columns:
  le = LabelEncoder()
  X_cat[i] = le.fit_transform(X_cat[i])
```

```python
X_cat.head()
```

```python
scaler = StandardScaler()
scaler.fit(X_num)
X_scaled = scaler.transform(X_num)
X_scaled = pd.DataFrame(X_scaled, index=X_num.index, columns=X_num.columns)
X = pd.concat([X_scaled, X_cat], axis=1)
y = df['price']
```

```python
X.head()
```

```python
X.isna().sum()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from math import sqrt
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model_lgb = lgb.LGBMRegressor()
```

```python
param={
    'num_leaves':[32,64],
    'min_data_in_leaf':[1,5,10],
    'colsample_bytree':[0.8,1],
    'n_estimators':[50, 100]
}
```

```python
GS_LGB = GridSearchCV(model_lgb, param, cv=10, n_jobs=n_thread)
GS_LGB.fit(X_train, y_train)
```

```python
GS_LGB.best_params_
```

```python
pred = GS_LGB.predict(X_test)
print(mean_absolute_error(y_test, pred))
print(sqrt(mean_squared_error(y_test, pred)))
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, pred)
print(accuracy)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBClassifier, XGBRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import label_ranking_average_precision_score

le = LabelEncoder()
y_train = le.fit_transform(y_train)

# Инициализируем модель XGBoost
model = XGBRegressor()

# Определим диапазоны гиперпараметров для GridSearchCV
params = {
    'learning_rate': [0.2, 0.3],
    'n_estimators': [50, 70, 100],
    'max_depth': [10, 15, 20]
}
#model.fit(X_train, y_train)
# Создадим экземпляр GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=3)

# Выполним поиск по сетке
grid_search.fit(X_train, y_train)

# Получим лучшую модель после GridSearchCV
best_model = grid_search.best_estimator_

# Получим количество деревьев (n_estimators) лучшей модели
best_n_estimators = best_model.get_params()['n_estimators']
print(f'Количество деревьев в лучшей модели: {best_n_estimators}')

# Сделаем предсказания на тестовом наборе данных
y_pred = best_model.predict(X_test)

# Оценим точность модели
accuracy = r2_score(y_test, y_pred)
print(f"Точность модели: {accuracy}")

```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
n_neighbors_values = [9, 11, 13]
weights_values = ['uniform','distance']
p = [1,2]
model = KNeighborsClassifier()
param_grid = {'n_neighbors': n_neighbors_values, 'weights': weights_values, "p": p}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring= 'accuracy', cv=3)
grid_search.fit(X_train, y_train)
results = grid_search.cv_results_
best_params = grid_search.best_params_
best_accuracy = grid_search.best_score_ 
best_knn_model = grid_search.best_estimator_
y_pred = best_knn_model.predict(X_test)
print(f' KNN: ')
print(f'Наилучшие параметры: {best_params}')
print(f'Соответствующая точность: {r2_score(y_test,y_pred):.4f}')
```
